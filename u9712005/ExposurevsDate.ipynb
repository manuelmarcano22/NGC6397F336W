```python
# Make a Plot of the Exposure time and the Observation Time
```

```python
import glob
from astropy.io import fits
from astropy.time import Time
import matplotlib.pyplot as plt
import numpy as np

from bokeh.plotting import output_notebook, figure, show
from bokeh.models import HoverTool, tools,ColumnDataSource
output_notebook()
```

```python
lista = glob.glob('Originals/u*c0m.fits')
dates = []
times = []
exptime = []
for i in lista:
    fit= fits.open(i)
    dates.append(fit[0].header['DATE-OBS'])
    times.append(fit[0].header['TIME-OBS'])
    exptime.append(fit[0].header['EXPTIME'])
    
```

```python
datesobs = [['{}T{}'.format(i,j)] for i,j in zip(dates,times)]
print(datesobs[0:3])
```

```python
d = np.array(datesobs)
t = Time(d)
```

```python
data = {'time':t.datetime64,
       'y':exptime,
       'names':datesobs}

source = ColumnDataSource(data=data)

p = figure(plot_width=900, plot_height=500, title='a',active_drag='pan', active_scroll='wheel_zoom',
          x_axis_type='datetime',y_axis_label='ExpTime',x_axis_label='Date_Obs')


#Tool to get wavelength
hover2 = HoverTool(
        tooltips=[
            ('Date', '(@names)')
        ]
    )

        
p.add_tools(hover2)



# add a circle renderer with a size, color, and alpha
p.circle('time','y', source=source)

p.add_tools(hover2)

show(p)
```
