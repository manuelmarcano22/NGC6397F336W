```python
import pandas as pd
```

```python
data = pd.read_csv('gaianearu18.tsv',sep='\t',skiprows=103)
```

```python
data.head()
```

```python
data = data.drop([0,1])
```

```python
regionfile = 'regiongaia.reg'
with open(regionfile,'w') as fileone:
    for raa,decc,idstar in zip(data['_RAJ2000'],data['_DEJ2000'],data['Source']):
        text = f'fk5;circle({raa},{decc},0.1") # color=red text = {{{idstar}}}\n'
        fileone.write(text)
```
