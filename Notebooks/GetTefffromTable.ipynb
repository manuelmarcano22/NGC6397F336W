#  NGC 6397: The metallicity trend along the isochrone revisited 

They published a T_eff of some stars. Get the Ra and dec to see if U18 is there

```python
import pandas as pd
```

```python
data=pd.read_csv('TeffRashietal2020.tsv',sep='\t',skiprows=61)
data = data.drop([0,1])
```

```python
data.head()
```

```python
data['ID'] = pd.to_numeric(data['ID'])
```

```python
regionfile = 'regionTeff.reg'
with open(regionfile,'w') as fileone:
    for raa,decc,idstar in zip(data['RAJ2000'],data['DEJ2000'],data['ID']):
        text = f'fk5;circle({raa},{decc},0.1") # color=green text = {{{idstar}}}\n'
        fileone.write(text)
```
