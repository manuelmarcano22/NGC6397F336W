```python
info2 = []
for name,ra,dec in info:
    ra1,ra2,ra3 = ra.split()
    #rastring.append('{}:{}:{}'.format(ra1,ra2,ra3))
    rastring = '{}:{}:{}'.format(ra1,ra2,ra3)
    dec1,dec2,dec3 = dec.split()
    decstring = '{}:{}:{}'.format(dec1,dec2,dec3)
    info2.append((name,rastring,decstring))
```
